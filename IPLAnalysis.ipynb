```python
vk = match[match['batsman']=='V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
def runs_scored(batsman_name):
    batsman_name = match[match['batsman']==batsman_name]
    return batsman_name.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1).index[0]
```

```python
runs_scored('MS Dhoni')
```
